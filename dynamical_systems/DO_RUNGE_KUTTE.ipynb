```python
import matplotlib.pyplot as plt
```

```python
def Lk(f,t,y,h):
    L=[]
    L.append(f(t,y))
    L.append(f(t+h/2,y+h*L[0]/2))
    L.append(f(t+h/2,y+L[1]/2))
    L.append(f(t+h,y+h*L[2]))
    return L

def rk(f,h,t0,y0,tmax):
    t=t0
    y=[]
    time=[]
    y.append(y0)
    i=1
    L=[4]
    while (t<=tmax):
        L=Lk(f,t,y[i-1],h)
        y.append( y[i-1]+ h/6*( L[0]+2*L[1]+2*L[2]+L[3] ) )
        time.append(t)
        i+=1
        t+=h
    
    time.append(t)
    return y,time;
```

```python
def f(t,y):
    return 1+t-y
```

```python
y,t=rk(f,0.1,0,0,1)
plt.scatter(y, t)
```

```python
y,t=rk(f,0.25,0,0,1)
plt.scatter(y, t)
plt.plot(y,t)
```
